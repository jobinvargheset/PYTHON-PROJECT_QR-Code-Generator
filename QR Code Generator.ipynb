```python
import qrcode

def generate_qr_code(data, filename):
    """Generates a QR code image and saves it to a file.

    Args:
        data (str): The data to encode in the QR code.
        filename (str): The filename to save the image to (e.g., 'qrcode.png').
    """
    img = qrcode.make(data)
    img.save(filename)
    print(f"QR code saved to {filename}")

# Example usage:
data_to_encode = "https://www.amazon.com"  # Replace with your desired data
filename = "my_qrcode.png"  # Replace with your desired filename
generate_qr_code(data_to_encode, filename)
```
